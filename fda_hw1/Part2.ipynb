# Google Play Store Apps

## First let us preview the dataset

```python
import pandas as pd
import locale
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
%matplotlib inline

locale.setlocale( locale.LC_ALL, 'en_US.UTF-8' )

pd_data = pd.read_csv('googleplaystore.csv')
pd_data = pd_data.drop(index=10472)
pd_data = pd_data.drop_duplicates(subset=['App'], keep='first')

# Make install an integer
pd_data.Installs = pd_data['Installs'].apply(lambda x:locale.atoi(x.replace('+', '')))


pd_data[:1000]
```

## Free vs Paid

```python
appType = pd_data.groupby(['Type'])['Type'].count()
appType.plot.pie(y='Type',figsize=(10, 10))
plt.show()
```

## Installations and App Creations based on Category

```python
categoryInstall = pd_data.groupby(['Category'])['Installs'].sum()
categoryCount = pd_data.groupby(['Category'])['App'].count()

categoryAppInstall = pd.concat([categoryCount, categoryInstall], axis=1)

fig = plt.figure(figsize=(15,10)) # Create matplotlib figure

ax = fig.add_subplot(111) # Create matplotlib axes
ax2 = ax.twinx() # Create another axes that shares the same x-axis as ax.

width = 0.4

# Creat the overlap grouped twin axis plot
categoryAppInstall.Installs.plot(kind='bar', color='red', ax=ax, width=width, position=1)
categoryAppInstall.App.plot(kind='bar', color='blue', ax=ax2, width=width, position=0)

# legend will overlap, separate them
ax.legend(loc='upper left')
ax2.legend(loc='upper right')

# grid off
ax.grid(b=False)
ax2.grid(b=False)

# labels
ax.set_ylabel('App Installed')
ax2.set_ylabel('App Created')

plt.show()
```

The bar chart above is not a best way to show the relation between apps created and apps install, so I try to redo it with the install/app ratio

### It is a good practice to know the apps in play store how recent they update their apps. This is an important index for app development and apps' inactivities.

Lets see among these apps, the distribution of apps being updated over the date time.

```python
import datetime
import time

# make last update object to 'date' object
type(pd_data['Last Updated'].iloc[0])

# we will see csv will import all strings
# eg January 7, 2018

prevDate = pd_data['Last Updated'][:5]
print(prevDate)

# Now transform the date in str to date in object
pd_data['Last Updated'] = pd_data['Last Updated'].apply(lambda x:datetime.datetime.strptime(x,'%B %d, %Y'))
time_data = pd_data['Last Updated'].apply(lambda x:time.localtime(int(time.mktime(x.timetuple()))).tm_year)


nowDate = pd_data['Last Updated'][:5]
print(nowDate)
```

```python
dateCount = time_data.value_counts()
dateCount = dateCount.sort_index()
plt.subplots(figsize=(15,7))
barData = plt.bar(dateCount.index, dateCount)
plt.grid(b=False)
```

### Now take a look at top 10 most review app in play store

```python
pd_data['Reviews'] = pd_data['Reviews'].apply(lambda x:locale.atoi(x))
topReview = pd_data.sort_values(by=['Reviews'],ascending=False)[:10]
topReview
```

### What is the relationship between these columns

After viewing the most reviewed app, you must be wondering how other columns are related to those top apps
Now it's time to construct a correlation heatmap for the apps.

```python
# Make all columns a number

# Make a copy of original da
data = pd_data.copy()
data['Rating'] = data['Rating'].apply(lambda x:float(x))

def to_size(x):
    if x=='Varies with device':
        return 0
    elif x[-1] == "M":
        x = x.replace('M','')
        return float(x)*1000
    elif x[-1] == "k":
        x = x.replace('k','')
        return float(x)
    return 0

data['Size'] = data['Size'].apply(to_size)

def to_type(x):
    if x == 'Free':
        return 0
    else:
        return 1
    
data['Type'] = data['Type'].apply(to_type)
data['Last Updated'] = data['Last Updated'].apply(lambda x:x.timestamp())
data
    
```

### Builiding correlation between some columns

```python
data = data[['Rating','Size','Type','Installs','Reviews','Last Updated']]
correlation = data.corr()
print(data)
plt.figure(figsize=(15, 8))
heat_ax = sns.heatmap(correlation, xticklabels=correlation.columns, yticklabels=correlation.columns, annot=True)
```
