```python
import pandas as pd

pd_data = pd.read_csv("Reviews.csv")
pd_data = pd_data[:10000]
# pd_data.sort_values(by=['ProfileName','UserId'])
pd_data
```

```python
q1data = pd_data.copy()
score_mean = q1data.groupby(['ProfileName', 'UserId'])['Score'].mean()
score_mean = score_mean.sort_values()
# score_mean
```

```python
count_sum = q1data.groupby(['ProfileName', 'UserId'])['UserId'].count()
count_sum = count_sum.sort_values(ascending=False)
# count_sum
```

```python
df = pd.concat([count_sum,score_mean], axis=1)
df.columns = (['Score count', 'Score mean'])
df = df.sort_values(by='Score count', ascending=False)
df = df.swaplevel()[:10]
df
```

```python
# the most reviewed user
the_user = df.iloc[0]
username = the_user.name[0]
user_reviews  = pd_data[pd_data['UserId']==username]
```

```python
score_group = user_reviews.groupby(['Score'])['Score'].count()
score_group
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
colors = ["blue", "orange", "green", "red", "purple"]
score_group.plot.bar(color=colors)
plt.show()
```

```python
# from datetime import datetime
import time
time_data = pd_data['Time']
time_data = time_data.apply(lambda x:time.localtime(x).tm_year)
```

```python
%matplotlib inline
count = time_data.value_counts()
count = count.sort_index()
count.plot.bar()
```

```python
import seaborn as sns; sns.set()
correlation_data = pd_data.drop(['ProductId', 'Summary','Text', 'UserId', 'ProfileName'], axis=1)
correlation = correlation_data.corr()
print(correlation)
heat_ax = sns.heatmap(correlation, xticklabels=correlation.columns, yticklabels=correlation.columns, annot=True)
```

```python
import math
helpful_data = pd_data[pd_data.HelpfulnessNumerator <= pd_data.HelpfulnessDenominator]
helpful = helpful_data.HelpfulnessNumerator / helpful_data.HelpfulnessDenominator
helpful = helpful.apply(lambda x: -1 if math.isnan(x) else x)
helpful.hist(grid=True)
plt.show()
```
