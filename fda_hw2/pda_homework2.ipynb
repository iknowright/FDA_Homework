### import packages

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.externals.six import StringIO   
from sklearn.tree import export_graphviz
import pydotplus
import numpy as np
```

    - From 2k - 500k
- Sex

    - From degree 0 to 5

### Construct Random Sample

```python
pd_data = pd.DataFrame(np.random.randint(16, 80, size=(1000,1)), columns = ['Age'])
pd_data['Income'] = np.random.normal(35000, 15000, len(pd_data))
pd_data
```

### Make All Income Positive 

```python
def abs_neg(x):
    if x < 0:
        return int(abs(x))
    return int(x)
pd_data.Income = pd_data.Income.apply(abs_neg)
pd_data.head()
```

### Add Expenses

```python
pd_data['Expenses'] = np.random.normal(30000, 17500, len(pd_data))
pd_data.Expenses = pd_data.Expenses.apply(abs_neg)
pd_data.head()
```

### Add Rest of The Columns

```python
pd_data['Online Shopping Experience'] = np.random.randint(0, 5, len(pd_data))
pd_data['Sex'] = np.random.choice(list({'Male', 'Female'}), len(pd_data))
pd_data['Debit Card Holder'] = np.random.choice(list({'0', '1'}), len(pd_data))
pd_data
```

## Prepare function to evaluate my 'right' result

![](right_tree.png)


```python
def apply_result(x):
        if x['Online Shopping Experience'] == 0 or x['Online Shopping Experience'] == 1:
            if x.Income > 100000:
                return 'YES'
            else:
                if x.Age < 25:
                    return 'NO'
                else:
                    if x.Expenses > 50000:
                        return "YES"
                    else:
                        return "NO"
        elif x['Online Shopping Experience'] == 2 or x['Online Shopping Experience'] == 3:
            if x.Income > 75000:
                return "YES"
            else:
                if x.Age < 25:
                    if x.Sex == 'Male':
                        if x['Debit Card Holder'] == 1:
                            return "NO"
                        else:
                            if x.Expenses > 35000:
                                return "YES"
                            else:
                                return "NO"
                    else:
                        if x.Expenses > 35000:
                            return "YES"
                        else:
                            return "NO"
                else:
                    if x.Income > 45000:
                        return "YES"
                    else:
                        if x.Age>60:
                            return "NO"
                        else:
                            return "YES"              
        elif x['Online Shopping Experience'] == 4 or x['Online Shopping Experience'] == 5:
            return "YES"
        
pd_data['Credit Card'] = pd_data.apply(apply_result, axis=1)
pd_data
```

### Deal with sex and result

```python
def do_sex(x):
    if x=="Male":
        x = 1
    elif x=="Female":
        x = 0
    return x

def do_credit(x):
    if x=="YES":
        x = 1
    elif x=="NO":
        x = 0
    return x
pd_data['Credit Card'] = pd_data['Credit Card'].apply(do_credit)
pd_data.Sex = pd_data.Sex.apply(do_sex)    
pd_data.head()
```

## Seperate data into training set and testing(validation) set

```python
df_train = pd_data[:700]
df_train.head()
```

```python
df_test = pd_data[700:]
df_test.head()
```

## seperate attributes and answer

```python
y = df_train['Credit Card'].values
df_train = df_train.drop('Credit Card', 1)

y
```

```python
dtree=DecisionTreeClassifier(max_depth=4)
dtree.fit(df_train,y)

dot_data = StringIO()
export_graphviz(dtree, 
                out_file=dot_data,  
                filled=True, 
                feature_names=list(df_train),
                class_names=['non-holder','holder'],
                special_characters=True)

graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
graph.write_pdf("tree.pdf")
```

# Apply to testing data and then calculate accuracy.

```python
y_test = df_test['Credit Card'].values
X_test = df_test.drop('Credit Card', 1)

y_predict = dtree.predict(X_test)

y_predict
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_predict)
```
